```python
# Setup detectron2 logger
import detectron2

import numpy as np 
import os, json, cv2, random
from google.colab.patches import cv2_imshow

from detectron2 import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog, DatasetCatalog
```

```python
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml"))
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST= 0.5  # set threshold for this model
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml")
predictor = DefaultPredictor(cfg)
```

```python
from os import listdir
from os.path import isfile, join
import numpy
import re

numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

def getDetections(mypath, filename, target_class):
  # onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
  # images = numpy.empty(len(onlyfiles), dtype=object)
  # for n in range(0, len(onlyfiles)):
  #   images[n] = cv2.imread( join(mypath,onlyfiles[n]) )
  onlyimagesfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f))]
  images = numpy.empty(len(onlyimagesfiles), dtype=object)
  for n, infile in enumerate(sorted(onlyimagesfiles, key=numericalSort)):
    images[n] = cv2.imread( join(mypath,infile))
  
  f = open(filename, 'w')
  for i, image in enumerate(images):
    output = predictor(image)
    instance = output['instances']
    for j, cls in enumerate(instance.pred_classes.cpu().numpy()):
      x1 = instance.pred_boxes[j].tensor[0,0].data.cpu().numpy() # x1
      y1 = instance.pred_boxes[j].tensor[0,1].data.cpu().numpy() # y1
      x2 = instance.pred_boxes[j].tensor[0,2].data.cpu().numpy() # x2
      y2 = instance.pred_boxes[j].tensor[0,3].data.cpu().numpy() # y2
  
      if (cls == target_class):
        f.write(str(i) + ",-1,")
        f.write(str(x1) + "," + str(y1) + ',' + str(x2 - x1) + ',' + str(y2 - y1) + ',')
        score = instance.scores[j].cpu().numpy()
        f.write(str(score) + ',')
        f.write('-1,-1,-1' + '\n')

```

```python
import csv
import cv2

def visualizeTracks(img_dirs, track_dirs, classes):
  for i, dir in enumerate(img_dirs):
    # get images
    onlyimagesfiles = [ f for f in listdir(dir) if isfile(join(dir,f))]
    images = numpy.empty(len(onlyimagesfiles), dtype=object)
    for n, infile in enumerate(sorted(onlyimagesfiles, key=numericalSort)):
      images[n] = cv2.imread( join(dir,infile))

    #get track info
    onlytrackfiles = [f for f in listdir(track_dirs[i]) if isfile(join(track_dirs[i],f))] # different tracking files for each class
    for n in range(0, len(onlytrackfiles)):
      cls = int(onlytrackfiles[n].split("_")[1].split('.')[0])
      with open(join(track_dirs[i], onlytrackfiles[n]), newline = '') as csvfile:
        reader = csv.reader(csvfile, delimiter = ',')
        for row in reader:
          framenum = int(row[0])
          start = (int(float(row[2])), int(float(row[3])))
          end = (int(float(row[4]) + float(row[2])), int(float(row[5]) + float(row[3]))) #x,y,w,h to x1,y1,x2,y2
          cv2.rectangle(images[framenum], start, end, (255,255,255), 1)

          classname = classes[cls]
          cv2.putText(images[framenum], classname + " " + str(row[1]), start, 0, 0.5, (0,0,255), 1)


    fourcc = cv2.VideoWriter_fourcc('M','J','P','G')
    size = (images[0].shape[1], images[0].shape[0])
    out = cv2.VideoWriter(dir.split('/')[2] + '.avi',fourcc, 10, size)
    for image in images:
      out.write(image)
    out.release()
    cv2.destroyAllWindows()

    
```

```python
img_dirs = ['testing/image_02/0007'] #, 'testing/image_02/0018', 'testing/image_02/0027'] 
track_dirs = ['tracks/0007'] #, 'tracks/0018', 'tracks/0027']
classes = ['person', '', 'car', '', '', 'bus', '', 'truck']
visualizeTracks(img_dirs, track_dirs, classes)
```
